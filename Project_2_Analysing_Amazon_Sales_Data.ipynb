We will create new features from these:

1.ship day name - this is basically the day of the week when order took place.

2.Shipping duration - Total days spent for shipping

```python
# create 'Shipping duration' column using np.timedelta64(1,'D')
df['Shipping duration'] = (df['Ship Date'] - df['Order Date'])/np.timedelta64(1,'D')

# create 'Ship day name' column using 'Ship Date' column using day_name()
df['Ship day name'] = df['Ship Date'].dt.day_name()

print(df.info())
df.head(1)
```

```python
# as the profit depends on the difference of unit selling price and unit cost price, we can create a new column
# for this difference as Unit Profit

df['Unit Profit'] = round(df['Unit Price'] - df['Unit Cost'],2)

# using round command to round off it to two decimal places
```

```python
df.isnull().sum()
```

From the table we observe that we have no missing values

We have 'Order Priority' column given in terms of letters H, C, L, M

Assuming H - High, C - Critical, L - Low, M- Medium priorities

We will replace the column with these strings

```python
def order_priority(x):
  if x=="H":
    return "High"
  elif x=="C":
    return "Critical"
  elif x=="L":
    return "Low"
  elif x=="M":
    return "Medium"
df['Order Priority'] = df['Order Priority'].apply(order_priority)
```

### Since we are trying to understand the distribution of continuous numerical variables, we will be using

- histograms
- box plots

Below we have used a for loop to loop through all the continuous variables and then draw histograms and box plots for each of them at each iteration.

```python
#for loop for continuous_columns variable
for feature in continuous_columns:
    import warnings
    warnings.filterwarnings('ignore') #Ignore warnings
    # We created two variables, fig and ax
    # These two variables now hold the two core objects used for all types of plotting operations.
    fig,ax = plt.subplots(1,2,figsize=(12,6))
    #First object fig, short for figure, imagine it as the frame of your plot.
#     You can resize, reshape the frame but you cannot draw on it.
#     On a single notebook or a script, you can have multiple figures.
#     Each figure can have multiple subplots. Here, subplot is synonymous with axes.
#     The second object, ax, short for axes, is the canvas you draw on.
#     Or rephrasing, it is the blank sheet you can plot and hold your data.
#     An axes object can only belong to one figure.
    ax[0].hist(df[feature])
    ax[0].set_title('histogram of column values in '+feature)
    sns.boxplot(df[feature],ax=ax[1])
    #using ax2.set_title for box plot
    ax[1].set_title('box plot of column values in '+feature)
    # seaborn style setting
    sns.set(font_scale=1.1)
    # matplotlib command for displaying plots
    plt.show()
```

## Bivariate Analysis
We need to find out Sales-trend -> month wise , year wise , yearly_month wise

### 1(A).Monthly Sales And Profit

```python
#Define the desired order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

#Convert the 'Order Date' column to Categorical data type with the defined month order
df['Order Month'] = pd.Categorical(df['Order Date'].dt.strftime('%B'), categories=month_order, ordered=True)

#Group by month and calculate total sales
monthly_sales = df.groupby('Order Month')['Total Revenue'].sum().round(2)

print("Monthly Sales:")
print(monthly_sales)
```

```python
#Visualise using Line Chart
plt.figure(figsize=(12, 8))
plt.plot(monthly_sales.index, monthly_sales.values, marker='d', linestyle='-', color='yellow', linewidth=3)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.title('Visualise the Monthly Sales', fontsize=16, fontweight='bold')
plt.xticks(monthly_sales.index)

#Show the data month-wise on the graph
for x, y in zip(monthly_sales.index, monthly_sales.values):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10)

plt.legend(['Monthly Sales'], loc='upper right')
plt.gca().set_facecolor('white')

plt.show()
```

### 1(B).Units Sold vs Month

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
Monthly_Sale = df[["Month", "Units Sold"]].groupby("Month").sum()
Monthly_Sale = Monthly_Sale.sort_values("Month", key=lambda x: pd.Categorical(x, categories=month_order, ordered=True))
Monthly_Sale
```

```python
#Visualize Monthly Sales Using Line chart
plt.figure(figsize=(12, 8))
plt.plot(Monthly_Sale.index, Monthly_Sale.values, marker='o', linestyle='-', color='red', linewidth=3)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.title('Visualise the Monthly Units Sold', fontsize=16, fontweight='bold')
plt.xticks(Monthly_Sale.index)

#Show the data month-wise on the graph
for x, y in zip(Monthly_Sale.index, Monthly_Sale.values):
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10)

plt.legend(['Monthly Units Sold'], loc='upper right')
plt.gca().set_facecolor('white')

plt.show()
```

We can see highest units sold are in Jul month followed by Feb, and least in Aug, March and Dec

Now we can find out the trend for yearly_month wise, for that better we can make a heatplot using pivot table

### 3.Yearly-Monthly-Sales And Profit

```python
#Group by year and month and calculate total sales
year_month_sales = df.groupby([df["Year"], df["Month"]])['Total Profit'].sum()

print("\nYearly-Monthly Sales:")
Yearly_Month_Sales=pd.DataFrame(year_month_sales)
Yearly_Month_Sales
```

```python
pivot_table = Yearly_Month_Sales.pivot_table(values='Total Profit', index='Month', columns='Year', aggfunc='sum')
pivot_table=pivot_table.fillna(0)
pivot_table
```

```python
#Visualize the Yearly-Monthly-Profit using Pivot_table thorough Heatmap:
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='GnBu',annot=True)
plt.title('Total Profit (Yearly_Month_Wise) Heat Map',fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Month',fontsize=15)
plt.show()
```
